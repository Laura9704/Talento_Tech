### **Listas**

permiten generar registos en forma lineal(horizantal) en n posiciones de informacion, esta puede ser modifcado por diversos comandos.

```python
#listas

nombres={"ana","carme","luis"}
edades={35,50,16}
varios={"cc",True,"talla"}

print("\n\nlistas")
print(nombres)
print(edades)
print(varios)

#modifcar listas

mi_lista = [10, 20, 30]
print("\n\noriginal\n",mi_lista)
mi_lista[1] = 25
print(" modifcado por posicion\n",mi_lista)
mi_lista.insert(1, 20)
print("modifcado por comando insert",mi_lista)
mi_lista.append(20)
print("modifcado por comando appened",mi_lista)
mi_lista.remove(20) # otra forma es indicar la posicion dentro de la lista edades[2]
print("modificar eliminando datos con remove", mi_lista)
mi_lista.insert(1, 20)
mi_lista.pop(2)
print("modificar eliminando datos con pop e insertado un dato en la posicion del dato eliminado", mi_lista)
```
