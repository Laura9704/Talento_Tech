```python
import pandas as pd

x = [1,2,2,3,4,4,5,6]
y = [2,3,4,4,4,6,5,7]

data = {'x': x, 'y': y}

df = pd.DataFrame(data)

df["xy"] = df["x"] * df["y"]
df["x2"] = df["x"] ** 2

print(df)
```

```python
sumas = df.sum()

print(sumas)
```

## **Regresión lineal**


```python
n = len(df)

sum_x = df["x"].sum()
sum_y = df["y"].sum()
sum_xy = df["xy"].sum()
sum_x2 = df["x2"].sum()

m = (n*sum_xy - sum_x*sum_y) / (n*sum_x2 - sum_x**2)

print("Pendiente:", m)
```

```python
b=(sum_y/len(df))-m*(sum_x/len(df))

print("Intercepto:", b)
```
