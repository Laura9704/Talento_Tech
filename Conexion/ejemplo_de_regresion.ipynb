**Métricas de Evaluación para Regresión Lineal**

Para evaluar la precisión de un modelo de regresión lineal, se utilizan varias métricas. Las más comunes son:

**Coeficiente de Determinación (R-cuadrado):**

¿Qué es? Mide la proporción de la varianza en la variable dependiente que es predecible a partir de la(s) variable(s) independiente(s). Un valor de 1.0 indica que el modelo explica toda la variabilidad de la variable de respuesta alrededor de su media, mientras que un valor de 0 indica que el modelo no explica ninguna variabilidad.
Cómo interpretarlo: Un R-cuadrado más alto generalmente indica un mejor ajuste del modelo a los datos, pero hay que tener cuidado con el sobreajuste. No se ha calculado en el notebook, pero se puede añadir fácilmente.

**Error Cuadrático Medio (MSE - Mean Squared Error):**

¿Qué es? Es el promedio de los cuadrados de los errores. Los 'errores' son las diferencias entre los valores observados y los valores predichos por el modelo. Al elevar al cuadrado los errores, se penalizan más los errores grandes.
Cómo interpretarlo: Un MSE más bajo indica un mejor ajuste del modelo. Ya lo has calculado en tu notebook con mse = 0.443396.

**Raíz del Error Cuadrático Medio (RMSE - Root Mean Squared Error):**

¿Qué es? Es la raíz cuadrada del MSE. Es una métrica muy utilizada porque las unidades del RMSE son las mismas que las de la variable dependiente, lo que facilita su interpretación.
Cómo interpretarlo: Un RMSE más bajo indica un mejor ajuste. También lo has calculado en tu notebook con rmse = 0.66588.

**Error Absoluto Medio (MAE - Mean Absolute Error):**

¿Qué es? Es el promedio de los valores absolutos de los errores. A diferencia del MSE, no penaliza los errores grandes tan fuertemente.
Cómo interpretarlo: Un MAE más bajo indica un mejor ajuste del modelo. Lo calculaste en tu notebook con mae = 0.570755.
En resumen, tanto el MSE, RMSE como el MAE te dan una idea de la magnitud promedio de los errores del modelo, mientras que el R-cuadrado te indica qué tan bien tu modelo explica la variabilidad en tus datos.

```python
import matplotlib.pyplot as plt

# calcular residuos si no existen
df2["residuos"] = df2["y"] - df2["ypred"]
plt.figure(figsize=(10,6))
plt.scatter(df2["ypred"], df2["residuos"], color="blue")
plt.axhline(y=0, color="red", linestyle="--")
plt.xlabel("Valores Predichos (ŷ)")
plt.ylabel("Residuos (y - ŷ)")
plt.title("Gráfico de Residuos vs Valores Predichos")

plt.grid(True)

plt.show()

# Histograma de Residuos
plt.figure(figsize=(10, 6))
sns.histplot(df2["y-ypred"], kde=True)
plt.xlabel('Residuos')
plt.ylabel('Frecuencia')
plt.title('Histograma de Residuos')
plt.grid(True)
plt.show()
```

**Gráfico de Residuos vs. Valores Predichos**

Eje X (Valores Predichos ($\hat{y}$$\hat{y}$)): Muestra los valores que tu modelo de regresión lineal predijo para la variable y.
Eje Y (Residuos ($y - \hat{y}$$y - \hat{y}$)): Muestra la diferencia entre los valores reales de y y los valores predichos por el modelo. Un residuo positivo significa que el modelo subestimó el valor real, y un residuo negativo significa que lo sobreestimó.
¿Cómo se interpreta?

El objetivo de este gráfico es ver si hay algún patrón en los residuos. Idealmente, los puntos deberían estar dispersos aleatoriamente alrededor de la línea horizontal y=0, sin ningún patrón discernible (curva, forma de cono, etc.). Si ves un patrón, esto podría indicar que:

No linealidad: Si los residuos forman una curva, tu modelo lineal podría no ser el adecuado y una relación no lineal podría ajustarse mejor a los datos.
Heterocedasticidad: Si la dispersión de los residuos aumenta o disminuye a medida que aumentan los valores predichos (forma de embudo o cono), la varianza de los errores no es constante, lo cual viola uno de los supuestos de la regresión lineal.
Valores atípicos o influyentes: Puntos individuales muy alejados del resto pueden ser valores atípicos.
En tu gráfico, con un número limitado de puntos de datos, observamos una dispersión relativamente aleatoria alrededor de la línea y=0. No hay un patrón curvo evidente, lo que sugiere que el supuesto de linealidad podría ser aceptable para estos datos. Sin embargo, podría haber una ligera tendencia a una mayor dispersión (heterocedasticidad) en algunos rangos, pero con tan pocos puntos es difícil confirmarlo definitivamente. La línea roja punteada en y=0 nos ayuda a visualizar dónde deberían centrarse los residuos.

**Histograma de Residuos**

Este gráfico muestra la distribución de los residuos. En una regresión lineal ideal, los residuos deberían seguir una distribución aproximadamente normal con una media cercana a cero.
¿Cómo se interpreta?

Tu histograma muestra que los residuos están centrados alrededor de cero, lo cual es bueno. Con una muestra pequeña de 8 puntos, es difícil esperar una forma de campana perfecta, pero la distribución parece razonablemente simétrica en torno a cero. Si el histograma estuviera sesgado o tuviera múltiples picos, podría indicar problemas con los supuestos del modelo o que faltan variables importantes.

En resumen, ambos gráficos te dan una visión visual de cómo de bien se ajusta tu modelo a los datos y si cumple con los supuestos clave de la regresión lineal.

```python
import statsmodels.api as sm

# Q-Q plot de Residuos
plt.figure(figsize=(8, 6))
sm.qqplot(df2["y-ypred"], line='s') # 's' para la línea de 45 grados
plt.title('Q-Q Plot de Residuos')
plt.grid(True)
plt.show()
```

### Interpretación del Q-Q Plot de Residuos

*   En el Q-Q Plot, si los puntos se alinean estrechamente a lo largo de la línea recta de 45 grados, esto indica que los residuos siguen una distribución aproximadamente normal. Esto es deseable en la regresión lineal.
*   Si los puntos se desvían significativamente de la línea, especialmente en los extremos, sugiere que los residuos no son normales. Esto podría implicar que los supuestos del modelo no se cumplen o que hay valores atípicos que afectan la distribución de los residuos.

Con una cantidad pequeña de datos como la que tienes (8 puntos), puede ser difícil ver una alineación perfecta, pero te da una idea general de la distribución.

**Métricas de Evaluación para Regresión Lineal**

Las métricas de evaluación nos ayudan a entender qué tan bien nuestro modelo predice los valores de la variable dependiente y si cumple con los supuestos subyacentes.

**Coeficiente de Determinación (R-cuadrado):**

¿Qué es? Mide la proporción de la varianza en la variable dependiente (Y) que es predecible a partir de la(s) variable(s) independiente(s) (X). Un valor de 1.0 (o 100%) indica que el modelo explica toda la variabilidad, mientras que 0 indica que no explica ninguna.
Interpretación: Un R-cuadrado más alto generalmente significa un mejor ajuste. Sin embargo, añadir más variables siempre aumentará el R-cuadrado, incluso si las variables no son significativas. Por eso, a menudo se prefiere el R-cuadrado ajustado.
R-cuadrado Ajustado (Adjusted R-squared):

¿Qué es? Es una versión modificada del R-cuadrado que se ajusta por el número de predictores en el modelo. Penaliza la adición de variables predictoras que no mejoran significativamente el modelo.
Interpretación: Es más útil que el R-cuadrado simple cuando se comparan modelos con un número diferente de predictores. Un R-cuadrado ajustado más alto indica un mejor modelo.

**Error Cuadrático Medio (MSE - Mean Squared Error):**

¿Qué es? Es el promedio de los cuadrados de los errores (la diferencia entre el valor observado y el valor predicho). Al elevar al cuadrado los errores, se penalizan más los errores grandes.
Interpretación: Un MSE más bajo indica un mejor rendimiento del modelo. Sus unidades son el cuadrado de las unidades de la variable dependiente, lo que a veces dificulta su interpretación directa.
Raíz del Error Cuadrático Medio (RMSE - Root Mean Squared Error):

¿Qué es? Es la raíz cuadrada del MSE. Es la métrica más utilizada porque las unidades del RMSE son las mismas que las de la variable dependiente, lo que facilita su interpretación.
Interpretación: Un RMSE más bajo indica un mejor ajuste. Es sensible a valores atípicos debido a la penalización de errores grandes.
Error Absoluto Medio (MAE - Mean Absolute Error):

¿Qué es? Es el promedio de los valores absolutos de los errores. No penaliza los errores grandes tan fuertemente como el MSE/RMSE.
Interpretación: Un MAE más bajo indica un mejor ajuste. Sus unidades son las mismas que las de la variable dependiente, por lo que es fácil de entender. Es más robusto a valores atípicos que el RMSE.

**Error Porcentual Absoluto Medio (MAPE - Mean Absolute Percentage Error):**

¿Qué es? Mide el tamaño del error en términos porcentuales. Calcula el promedio de los valores absolutos de los errores divididos por los valores observados, multiplicado por 100.
Interpretación: Un MAPE más bajo indica un mejor modelo. Es útil cuando quieres un error que sea fácil de entender en un contexto de negocio (ej. 'el error promedio es del 5%'). Sin embargo, tiene problemas cuando los valores reales son cero o muy cercanos a cero.

**Mediana del Error Absoluto (MedAE - Median Absolute Error):**

¿Qué es? Es la mediana de los valores absolutos de los errores. A diferencia del MAE, no es un promedio, lo que lo hace muy robusto a valores atípicos.
Interpretación: Un MedAE más bajo indica un mejor ajuste. Es particularmente útil cuando tus datos contienen muchos valores atípicos o la distribución de errores es muy asimétrica.
Consideraciones Adicionales:
Además de las métricas numéricas, es crucial utilizar gráficas de diagnóstico como las que hemos generado (diagrama de dispersión con línea de regresión, gráfico de residuos, histograma de residuos y Q-Q plot de residuos) para evaluar visualmente si el modelo cumple con los supuestos de la regresión lineal (linealidad, independencia de errores, homocedasticidad y normalidad de los residuos).

La elección de la métrica dependerá del contexto y los objetivos específicos de tu análisis. Por ejemplo, si los errores grandes son especialmente costosos, el RMSE podría ser preferible. Si la robustez a valores atípicos es clave, MAE o MedAE pueden ser mejores.

Las gráficas de diagnóstico son esenciales para evaluar si tu modelo de regresión lineal cumple con los supuestos y si se ajusta bien a los datos. Las principales son:
